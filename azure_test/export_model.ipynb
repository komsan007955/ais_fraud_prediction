```python
import numpy as np
import pandas as pd

df = pd.read_csv(f"dataset/interactions_train.csv")
```

```python
df.drop("date", axis=1, inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
import pickle

X = df.drop("rating", axis=1).values
y = df["rating"].values

lr = LinearRegression()
model = lr.fit(X, y)

pickle.dump(model, open(f"model/model.pkl", "wb"))
```
